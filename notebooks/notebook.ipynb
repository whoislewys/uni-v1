```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Constant Sum Market Maker

```python
# Set constant k
k = 5

# Generate x values starting from 0
x = np.linspace(0, 5, 100)  # Reduced range to 5 since k=5

# constant sum invariant in plain english: liquidity equals reserves of x + reserves of y
# in algebraic terms, k = x + y
# k = x + y
# Calculate y values starting from 0
y = k - x

# Create plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y)
plt.title('Constant Sum Market Maker (x + y = k)')
plt.xlabel('Reserve of x')
plt.ylabel('Reserve of y')
plt.grid(True)
plt.xlim(0, 7)  # Set x axis to start at 0
plt.ylim(0, 7)  # Set y axis to start at 0
plt.show()
```

## Constant Product Market Maker

```python

# Set constant k
k = 5

# Generate x values starting from 0
x = np.linspace(0, 5, 100)  # Reduced range to 5 since k=5

# constant sum invariant in plain english: liquidity equals reserves of x + reserves of y
# in algebraic terms, k = x * y
# Calculate y values starting from 0
y = k / x

# Create plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y)
plt.title('Constant Product Market Maker (x * y = k)')
plt.xlabel('Reserve of x')
plt.ylabel('Reserve of y')
plt.grid(True)
plt.xlim(0, 7)  # Set x axis to start at 0
plt.ylim(0, 7)  # Set y axis to start at 0
plt.show()
```
